politics

```python
politicslink = []
url_head = 'https://news.naver.com/main/list.naver?mode=LS2D&sid1=100&mid=shm'

for date in range(20231101, 20231117):
  urldate = '&date=' + str(date)

  # 대통령실 / 국회정당 / 북한 / 행정 / 국방외교 / 정치일반
  for i in range(4, 10):
    url1 = '&sid2=26' + str(i)

    # 각 세부섹션의 페이지
    for j in range(1, 3):
      url2 = '&page=' + str(j)

      url = url_head + urldate + url1 + url2

      res = requests.get(url, headers = headers)
      soup = BeautifulSoup(res.text, 'html.parser')
      newslist = soup.select('dt.photo > a.nclicks\(pol\.2ndcont\)')

      for k in range(0, len(newslist)):
        eachlink = newslist[k].get('href')

        if eachlink not in politicslink:
          politicslink.append(eachlink)
```

```python
len(politicslink)
```

```python
politicstext = []
idx = 0
for eachlink in politicslink:
  res = requests.get(eachlink, headers = headers)
  soup = BeautifulSoup(res.text, 'html.parser')

  # 제목
  try:
    title = soup.select('h2.media_end_head_headline')[0].span.text

  except:
    continue

  # 제목 제외한 나머지 텍스트 부분
  try:
    text = soup.select('article#dic_area')[0]

  except:
    continue

  try:

    # 주석
    text.find('span').text
    annotation = text.find('span').text.replace('\n', '')

  except:
    continue

  # 본문 (body)
  span_tag = text.find('span')
  body = ''.join(span_tag.find_next_siblings(text=True)).replace('\n', '').replace('\t', '')


  # 주석이나 본문 제대로 못 가져오면 pass
  if annotation == '' or body == '' or body == ' ' or body == ' r_start // r_end //':
    continue

  # 이미지
  imglink = soup.select('img#img1')[0].get('data-src')
  urllib.request.urlretrieve(imglink, '/content/drive/MyDrive/NaverNews/politics/' + 'politics' + str(idx) + ".jpg")


  eachtext = {
      'idx' : idx,
      'title' : title,
      'annotation' : annotation,
      'body' : body,
      'link' : eachlink
  }

  politicstext.append(eachtext)

  idx += 1
```

```python
len(politicstext)
```

```python
df_politicstext = pd.DataFrame(politicstext)
```

```python
df_politicstext.to_excel('/content/drive/MyDrive/NaverNews/politics/politicstext.xlsx', index = False)
```
